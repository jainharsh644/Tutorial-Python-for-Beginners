```python
#PATTERN - 1
n = int(input("Enter the no of layers"))
for i in range(1, n+1):
    for j in range(1, i+1):
        print("*", end = "")
    print()  
```

```python
#PATTERN - 2
n = int(input("Enter the no of layers"))
k = 1
for i in range(1, n+1):
    for j in range(1, k+1):
        print("*", end = "")
    k = k+2
    print()
```
